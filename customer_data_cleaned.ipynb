```python
# Loading the dataset using pandas

import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Summary statistics

df.describe(include = 'all')
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Imputing missing values with the median of that category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Standardizing column names

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
df.columns
```

```python
# Creating a new column 'age_group'

labels = ['young adult', 'adult', 'middle aged', 'senior']

df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df.groupby('age_group')['age'].agg(['min', 'max', 'count'])
```

```python
df['frequency_of_purchases'].unique()
```

```python
# Creating a new column 'purchase_frequency_days' 

mapping = {'Fortnightly': 14, 'Weekly': 7, 'Annually': 365, 'Quarterly': 90, 'Bi-Weekly': 14,
       'Monthly': 30, 'Every 3 Months': 90}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping 'promo_code_used' column

df = df.drop('promo_code_used', axis = 1)
```
